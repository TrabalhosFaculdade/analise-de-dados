```python
import matplotlib.pyplot as plt
import numpy as np

# data
grupo_1 = np.array([27,22,29,21,19,33,16,20,24,27,28,19])
grupo_2 = np.array([12,12,15,9,20,18,17,14,14,2,17,19])
grupo_3 = np.array([18,4,22,15,18,19,22,12,12])

# labels e valores usados no plot
objetos = ('Grupo 1', 'Grupo 2', 'Grupo 3')
valores = [np.median(grupo_1), np.median(grupo_2), np.median(grupo_3)]

# distancia entre os elementos
y_pos = np.arange(len(objetos))

plt.bar(y_pos, valores, align = 'center')
plt.xticks(y_pos, objetos)
plt.ylabel('Média de árvores')
plt.title('Quantidade de árvores por grupo')
 
plt.show()

plt.boxplot(np.array([grupo_1, grupo_2, grupo_3]))
plt.show()
```

Grupo 1: nunca sofreram extração
Grupo 2: sofreram extração há um ano
Grupo 3: sofreram extração há mais de oito anos

Podemos concluir a partir dos gráficos mostrados que a extração de árvores 
possui efeito direto na contagem das mesmas.

O gráfico abaixo, com as médias de árvore para cada um dos grupos determinados, 
demonstra que a floresta está em seu melhor estado quando não tocada.

A partir disso, as áreas que não sofreram extração por maiores períodos de 
tempo (Grupo 3) foram aquelas que demonstraram estar em melhores condições em 
relação às regiões que sofreram extrações em períodos mais próximos (Grupo 1).
