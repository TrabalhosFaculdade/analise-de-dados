```python
import pandas as pd
import numpy as np
import pyplot as plt

# 0 - index
# 1 - one
# 2 - age
# 3-  systolic blood pressure

values = []
labels = []
with open("blood_pressure.txt") as file:
    for line in file:
        line_data = [int(x) for x in line.rstrip().split(":")]
        labels.append(line_data[0])
        form_data = {
                        "one":line_data[1],
                        "age":line_data[2],
                        "pressure":line_data[3]
                    }
        
        values.append(form_data)
        
df = pd.DataFrame(values, index = labels)

print("Média de idade:", df['age'].mean())
print("Média de pressão sistólica:", df['pressure'].mean())
print("Desvio padrão de idade:", df['age'].std())
print("Desvio padrão de pressão sistólica:", df['pressure'].std())
print("Correlação entre idade e pressão sistólica:", df['age'].corr(df['pressure']))

a = df['age'].corr(df['pressure']) * df['pressure'].std() / df['age'].std()
b = df['pressure'].mean() - a * df['age'].mean()
print("Inclinação da reta de regressão pressão sistólica x idade:", a)
print("Intercepto da reta de regressão:", b)
print("Equação da reta de regressão pressão sistólica x idade:", ("{0} * x + {1}".format(str(a),str(b))))
print("Pressão de uma pessoa de 100 anos:", a * 100 + b)

df.plot.scatter(x = 'age', y = 'pressure')

"""

Qual seria a pressão arterial predita de uma pessoa de 100 anos?
Gráfico de dispersão + reta de regressão
"""

df
```
